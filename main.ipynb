###### V37: Generators in Python

```python
def f():
    return 1
    return 2
    return 3
print(f())

def g():
    yield 1
    yield 2
    yield 3
print(g())

for n in g():
    print(n, end=' ')
print()


import string

def letters():
    for c in string.ascii_lowercase:
        yield c

for letter in letters():
    print(letter, end=' ')
print()

import itertools
sqrs = (x ** 2 for x in itertools.count(1))
print(sqrs)
for n in sqrs:
    print(n, end=' ')
    if n > 500:
        sqrs.close()
print()

def primes():
    yield 2
    cache = [2]
    for n in itertools.count(3, 2):
        is_prime = True
        for p in cache:
            if n % p == 0:
                is_prime = False
                break
        if is_prime:
            cache.append(n)
            yield n

for p in primes():
    print(p, end=' ')
    if p > 100:
        break
```

###### V38: Regular Expressions in Python

```python
import re

# Position: ^(start) $(end)
# Set of characters \s(white space) \d(digit) \w(a-z, A-Z, 0-9, _) [aeiou] [a-z]
# Quantifiers  ?(0 or 1) *(0+) +(1+) {n}(exactly n occurencies)

names = [
    'Karl Marx',
    'Adam Smith',
    'Rene Descartes',
    'Carl Gauss',
    'Caesar',
    'Brutus',
    'Mark Antony',
    'm!sha'
]

pattern = '^\w+\s+\w+$'
for name in names:
    if match := re.search(pattern, name):
        print(name, match, match.start(), match.end(), match.span())

print('-' * 30)
pattern = '^(\w+)\s+(\w+)$'
for name in names:
    if match := re.search(pattern, name):
        print(match.group(1), match.group(2))

print('-' * 30)
pattern = 'C\w*'
for name in names:
    if match := re.search(pattern, name):
        print(name)

print('-' * 30)
pattern = '^[a-zA-Z!]+$'
for name in names:
    if match := re.search(pattern, name):
        print(name)

print('-' * 30)
pattern = '[a-z]+'
for name in names:
    if matches := re.findall(pattern, name):
        print(matches)
```

###### V49: Decorators in Python

```python
def compose(f, g, x):
    return f(g(x)) 
compose(print, len, 'hello world')

import random

def random_power():
    def f(x):
        return x ** 2
    def g(x):
        return x ** 3
    def h(x):
        return x ** 4
    
    functions = [f, g, h]
    return random.choice(functions)

for _ in range(10):
    p = random_power()
    print(p(3), end=' ')
```

```python
import time

# def prime_factorization(n):
#     start = time.time()
#     factors = []
#     divisor = 2

#     while n > 1:
#         while n % divisor == 0:
#             factors.append(divisor)
#             n //= divisor
#         divisor += 1
    
#     end = time.time()
#     print(f'Δt = {end - start}')
#     return factors

def timer(f):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = f(*args, **kwargs)
        end = time.time()
        print(f'Δt = {end - start}')
        return result
    return wrapper
    

@timer
def prime_factorization(n):
    factors = []
    divisor = 2

    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    return factors

ints = [2 ** 20 + 1, 2 ** 23 + 1, 2 ** 29 + 1, 2 ** 32 + 1]

for n in ints:
    f = prime_factorization(n)
    print(f)
    print('-' * 30)
```

```python
import functools
print(dir(functools))

from functools import wraps, cache
def do_nothing(f):
    @wraps(f)
    def wrapper(*args, **kwargs):
        return f(*args, **kwargs)
    return wrapper

@do_nothing
def alpha(*args, **kwargs):
    """A function for viewing arguments."""
    print(f'Args = {args}')
    print(f'Kwargs = {kwargs}')

# alpha('a', 2, None, x=7, y=11, z=28)
print(alpha.__name__)
print(alpha.__doc__)



@cache
def fib(n):
    if n < 2:
        return 1
    return fib(n - 1) + fib(n - 2)

@timer
def fib_wrapper(n):
    return fib(n)

for i in range(30, 37):
    print(f'Fib({i}) is {fib_wrapper(i)}')
    print('-' * 30)
```

###### V50: AcyncIO, await, and async - Concurrency in Python

```python
import time

def brew_coffee():
    print('start brew_coffee()')
    time.sleep(3)
    print('end brew_coffee()')
    return 'Coffee ready'

def toast_bagel():
    print('start toast_bagel()')
    time.sleep(2)
    print('end toast_bagel()')
    return 'Bagel toasted'

start = time.time()
coffee = brew_coffee()
bagel = toast_bagel()
end = time.time()
elapsed = end - start
print(coffee)
print(bagel)
print(elapsed)
```

```python
import asyncio
import time

async def brew_coffee():
    print('start brew_coffee()')
    await asyncio.sleep(3)
    print('end brew_coffee()')
    return 'Coffee ready'

async def toast_bagel():
    print('start toast_bagel()')
    await asyncio.sleep(2)
    print('end toast_bagel()')
    return 'Bagel toasted'



start = time.time()
batch = asyncio.gather(brew_coffee(), toast_bagel())
coffee, bagel = await batch
end = time.time()
elapsed = end - start
print(coffee)
print(bagel)
print(elapsed)
```
